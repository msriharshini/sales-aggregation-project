```python
import pandas as pd

jan = pd.read_csv("jan.csv")
feb = pd.read_csv("feb.csv")
mar = pd.read_csv("mar.csv")

all_data = pd.concat([jan, feb, mar], ignore_index=True)
all_data.head()
```

```python
# Group by Region
grouped = all_data.groupby("Region")

# Calculate summary statistics
summary = grouped["Sales"].agg(["sum", "mean", "median"]).reset_index()

# Display result
summary
```

```python
summary.to_csv("sales_summary.csv", index=False)
```
